```python
# Import library
# To do Graph empty
import networkx as nx
G = nx.Graph()
# Others types of Graphs
#G = nx.DiGraph()
#G = nx.MultiGraph()
#G = nx.MultiDiGraph()
```

```python
# Toagregate a node
G.add_node(1)
# Drawing
nx.draw(G)
```

```python
# Draw 10 nodes and graph in a circle manner
H = nx.path_graph(10)
G.add_nodes_from(H)
nx.draw(G, pos=nx.circular_layout(G), node_color='r', edge_color='b')
```

```python
# Draw interactions (edges) between nodes 
G.add_edge(1, 2)
e = (2, 3)
G.add_edge(*e)
nx.draw(G, node_color='g', edge_color='b', with_labels=True)
```

```python
# Clean the graph
G.clear()
nx.draw(G)
```

```python
# To do Digraph and agregate nodes and edges
DG = nx.DiGraph()
DG.add_edge(2, 1)   # adds the nodes in order 2, 1
DG.add_edge(1, 3)
DG.add_edge(2, 4)
DG.add_edge(1, 2)
assert list(DG.successors(2)) == [1, 4]
assert list(DG.edges) == [(2, 1), (2, 4), (1, 3), (1, 2)]
DG.number_of_nodes(), DG.number_of_edges(), nx.draw_circular(DG, with_labels=True)
```

```python
# New Graph
G = nx.Graph()
G.add_edge(1, 2)              # default edge data=1
G.add_edge(2, 3, weight=0.9)  # specify edge data
nx.draw_networkx(G)
```

```python
# Add nodes with math function
import math
G.add_edge('y', 'x', function=math.cos)
G.add_node(math.cos)  # any hashable can be a node
nx.draw_networkx(G)
```

```python
# Add differents interactions
elist = [(1, 2), (2, 3), (1, 4), (4, 2)]
G.add_edges_from(elist)
elist = [('a', 'b', 5.0), ('b', 'c', 3.0), ('a', 'c', 1.0), ('c', 'd', 7.3)]
G.add_weighted_edges_from(elist)
nx.draw_networkx(G, node_color='g', edge_color='b')
```

```python
# Two forms to view the same Graph
import matplotlib.pyplot as plt
G = nx.cubical_graph()
subax1 = plt.subplot(121)
nx.draw(G)   # default spring_layout
subax2 = plt.subplot(122)
nx.draw(G, pos=nx.circular_layout(G), node_color='r', edge_color='b')
```

```python
# Features selection - some algorithms
# Centrality
G = nx.path_graph(4)
centrality = nx.eigenvector_centrality(G)
sorted((v, f"{c:0.2f}") for v, c in centrality.items()), nx.draw(G)
```

```python
# Clustering
G = nx.complete_graph(7)
print(nx.clustering(G, 0))
print(nx.clustering(G)),  nx.draw(G)
```

```python
# Similaritary
G = nx.cycle_graph(2)
#nx.simrank_similarity(G)
#nx.simrank_similarity(G, source=0)
nx.simrank_similarity(G, source=0, target=0),  nx.draw(G)
```

```python
# Lynk Google
G = nx.DiGraph(nx.path_graph(4))
pr = nx.pagerank(G, alpha=0.9)
pr,  nx.draw(G, with_labels=True)
```

```python
# Example 
G = nx.Graph()        # Empty Graph
# Add Nodes
G.add_node("Yeikop")
G.add_node("Hans Peludo")
G.add_nodes_from(["GIO", "Osito", "Marthica"])
 
#Añadir aristas
G.add_edge("Yeikop", "Hans Peludo")
G.add_edge("Yeikop", "GIO")
G.add_edges_from([("Hans Peludo", "GIO"), ("Hans Peludo", "Osito")])
G.add_edges_from([("Osito", "GIO"), ("Osito", "Marthica")])
nx.draw_networkx(G)
```

```python
# Describe
print(len(G.nodes))
print(len(G.edges))
print(G.nodes)
print(G.edges)
```

```python
# Assign features to nodes
G.nodes["Hans Peludo"]["SOMOS"] = 2
G.edges["Yeikop", "Hans Peludo"]["VIVA"] = "Cerveza"
G.edges["Yeikop", "GIO"]["VIVA"] = "Vino"
G.edges["Osito", "GIO"]["Web"] = "Estampillas"
G.edges["Osito", "Hans Peludo"]["Estamos Ahí"] = "Pagan sus facturas"
G.edges["Osito", "Marthica"]["Homcenter"] = "Jardinería"
print(G.nodes(data=True))
print(G.edges(data=True))
print(G["Yeikop"])
```
